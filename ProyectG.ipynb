<h4 Style="color:#A52A2A"> ¿Cual marca de carros tenemos mas en inventario?</h4>

```python
import matplotlib as mpl

mpl.rcParams["axes.spines.bottom"] = True
mpl.rcParams["axes.spines.left"] = False
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
```

```python
dfAutos.head(3)
```

```python
#Se agrupan y cuenta la cantidad de registros segun marca
dfGraf = dfAutos.groupby(['make']).count()
#Se selecciona unicamente el indice y la primera columna
dfGraf = dfGraf.iloc[:,:1]
#Se ordenan los datos de mayor a menor
dfGraf = dfGraf.sort_values('symboling' , ascending = False)
#Se muestra el DataFrame para graficar
dfGraf
```

```python
plt.rcdefaults()
fig, ax = plt.subplots(1, 1, figsize=(11,6.9), dpi=100)

#Datos para crear el grafico
eje_x = dfGraf.index.tolist()
eje_y =dfGraf['symboling'].tolist()
y_pos = np.arange(len(eje_x))

#Gereracion de colores para el grafico
bar_colors =['#0c247d' ,'#04e8e6','#04e8e6','#04e8e6','#04e8e6','#04e8e6',
             '#04e8e6','#04e8e6','#04e8e6','#04e8e6','#04e8e6','#04e8e6',
             '#04e8e6','#04e8e6','#04e8e6','#04e8e6','#04e8e6','#04e8e6',
             '#04e8e6','#04e8e6','#04e8e6','#ff9908']

#Creacion del grafico
ax.barh(y_pos, eje_y, align='center' , color = bar_colors)

#Formato del grafico
#-----Titulo y subtitulo del grafico
title_t = "Figura 1: Cantidad de Vehículos"
ax.set_title(title_t,
            loc = 'left',
            x = 0.001,
            y= 0.96,
            transform=fig.transFigure,
            fontsize=24,
            fontweight ="bold",
            fontfamily ="serif")

subtitle_t = "Total de vehículos agrupados segun su marca o fabricante"
plt.suptitle(subtitle_t,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.92,
            transform=fig.transFigure,
            fontsize=16,
            fontweight ="regular",
            fontfamily ="serif")

label = "Cantidad"
ax.set_xlabel(label,
             ha='left',
             fontsize=10,
             fontweight="regular",
             fontfamily="sans")

ax.xaxis.set_label_coords(x=0,
                         y =-0.07,
                         transform=ax.transAxes)
ax.tick_params(axis="both", 
              labelsize=12)

ax.set_yticks(y_pos, labels=eje_x)
ax.invert_yaxis()
plt.subplots_adjust(top=0.80)

plt.show()
```
